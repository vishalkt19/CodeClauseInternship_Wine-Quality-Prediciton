 ## PREPROCESSING OF DATA

### BINARIZATION OF QUALITY 

```python
df['quality'].unique()
```

```python
df['quality']=[1 if x>=7 else 0 for x in df['quality']]
```

```python
df['quality'].unique()
```

```python
df['quality'].value_counts()
```

```python
plt.figure(figsize=(6,6),dpi=150)
sns.catplot(x='quality', data = df, kind = 'count')
```

### HANDLING OF IMBALANCED DATA SET  

```python
from imblearn.over_sampling import SMOTE
```

```python
X = df.drop('quality',axis=1)
y = df['quality']
```

```python
X,y = SMOTE().fit_resample(X,y)
```

### TRAIN TEST SPLIT 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

### SCALING OF DATA 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

### DIMENSIONALITY REDUCTION  

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.90)
```

```python
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

## LOGISTIC REGRESSION 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
logreg_pred = logreg.predict(X_test)
logreg_acc = accuracy_score(logreg_pred, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, logreg_pred))
print("\n")
print(classification_report(y_test, logreg_pred))
```

```python
from matplotlib import style
style.use('ggplot')
```

```python
style.use('classic')
cm = confusion_matrix(y_test, logreg_pred, labels=logreg.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels=logreg.classes_)
disp.plot()
```

### GRID SEARCH

```python
param_grid_LOG={'C':[0.01, 0.1, 1, 10],'penalty' : ['l1', 'l2'],'l1_ratio':[0.1,0.5,1]}
grid_LOG = GridSearchCV(LogisticRegression(max_iter=10000), param_grid_LOG)
grid_LOG.fit(X_train, y_train)
```

```python
print("Best parameters:", grid_LOG.best_params_)
```

```python
y_pred_LOG = grid_LOG.predict(X_test)
```

```python
logreg_acc = accuracy_score(y_pred_LOG, y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
print(confusion_matrix(y_test, y_pred_LOG))
print("\n")
print(classification_report(y_test, y_pred_LOG))
```

## USER INPUT PREDICTION

```python
def predict_input():
    
    feature_info = {
        'fixed acidity': float,
        'volatile acidity': float,
        'citric acid': float,
        'residual sugar': float,
        'chlorides': float,
        'free sulfur dioxide': float,
        'total sulfur dioxide': float,
        'density': float,
        'pH': float,
        'sulphates': float,
        'alcohol': float
    }
    user_input = {}
    for feature_name, data_type in feature_info.items():
        value = input(f"Enter {feature_name}: ")
        user_input[feature_name] = data_type(value)
    user_df = pd.DataFrame([user_input])
    test = pca.transform(scaler.transform(user_df))
    p = rfc.predict(test)
    if(p[0]==0):
        print('Bad Quality')
    else:
        print('Good Quality')
```
